## Business Validation Analysis
El objetivo de esta sección es comprobar si los datos permiten responder a las preguntas de negocio definidas para el proyecto.

En concreto, se analizarán diferencias entre regiones, almacenes y proveedores para validar la viabilidad del proyecto de optimización de inventario y procurement.

```python
#Ventas por region
df.groupby("Region")["Units_Sold"].sum()
```

Se observan diferencias en el volumen de ventas entre regiones, lo que permitirá realizar análisis comparativos y evaluar el comportamiento de la demanda en distintas zonas geográficas.

```python
#Inventario medio por almacen
df.groupby("Warehouse_ID")["Inventory_Level"].mean().sort_values(ascending=False)
```

El nivel medio de inventario varía entre almacenes, lo que sugiere posibles diferencias en la gestión del stock y en la estrategia de aprovisionamiento.

```python
#Lead time medio por proveedor
df.groupby("Supplier_ID")["Supplier_Lead_Time_Days"].mean().sort_values(ascending=False)
```

Los proveedores presentan diferentes tiempos de suministro, lo que puede afectar a las decisiones de reposición y al riesgo asociado al inventario.

```python
#Crear Procurement Spend
df["Procurement_Spend"] = (
    df["Order_Quantity"] *
    df["Unit_Cost"]
)
```

```python
#Gasto por proveedor
df.groupby("Supplier_ID")["Procurement_Spend"].sum().sort_values(ascending=False)
```

El gasto en compras no se distribuye de forma uniforme entre proveedores. Esto permitirá identificar proveedores estratégicos y analizar posibles dependencias.

```python
#Error de forecast 
df["Forecast_Error"] = (
    df["Units_Sold"] -
    df["Demand_Forecast"]
)
```

```python
df["Forecast_Error"].describe()
```

El análisis del error de forecast permitirá evaluar la precisión de la previsión de demanda y detectar posibles desviaciones entre la demanda estimada y la demanda real.
