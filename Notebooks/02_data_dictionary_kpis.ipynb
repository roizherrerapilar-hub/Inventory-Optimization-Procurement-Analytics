# 02 - Diccionario de Datos y Definición de KPIs

El objetivo de este notebook es documentar las variables disponibles en el dataset, definir las métricas de negocio y establecer las preguntas que guiarán el análisis.


## Business Case

Una empresa desea mejorar la gestión de inventario y optimizar sus decisiones de aprovisionamiento.

Para ello, se analizarán los niveles de inventario, la demanda, los puntos de reposición y el rendimiento de los proveedores con el objetivo de:

- Mejorar la eficiencia del inventario.
- Reducir costes de almacenamiento.
- Optimizar las decisiones de compra.
- Mejorar la visibilidad de la cadena de suministro.
- Analizar el impacto de factores externos como el tipo de cambio.

## Preguntas de Negocio

### Inventario

- ¿Qué productos mantienen mayores niveles de inventario?
- ¿Existen productos con exceso de stock?
- ¿Qué productos requieren reposición con mayor frecuencia?
- ¿Cómo se comporta la demanda frente al inventario disponible?

### Procurement

- ¿Qué proveedores generan mayor volumen de compras?
- ¿Qué proveedores presentan mayores lead times?
- ¿Qué proveedores concentran el mayor gasto?

### Almacenes

- ¿Qué almacenes gestionan mayores niveles de inventario?
- ¿Qué almacenes registran mayores ventas?

### Forecast

- ¿Cuál es la diferencia entre la demanda prevista y la demanda real?

### Factores Externos

- ¿Cómo podría afectar la volatilidad del tipo de cambio a los costes de aprovisionamiento?

## KPIs de Inventario

### Inventory Value

Valor económico del inventario disponible.

Fórmula:

Inventory Level × Unit Cost

---

### Inventory Coverage

Número de días que el inventario actual podría cubrir la demanda prevista.

Fórmula:

Inventory Level ÷ Demand Forecast

---

### Replenishment Frequency

Frecuencia con la que se generan pedidos de reposición.

Fórmula:

Número de registros con Order Quantity > 0

## KPIs de Procurement

### Procurement Spend

Valor total de las compras realizadas.

Fórmula:

Order Quantity × Unit Cost

---

### Average Supplier Lead Time

Tiempo medio de suministro por proveedor.

Fórmula:

Promedio de Supplier Lead Time Days

---

### Spend by Supplier

Distribución del gasto por proveedor.

## KPIs de Forecast

### Forecast Error

Diferencia entre demanda real y demanda prevista.

Fórmula:

Units Sold - Demand Forecast

---

### Forecast Accuracy

Nivel de precisión de la previsión de demanda.
